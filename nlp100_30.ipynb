** このファイルを用いて，以下の問に対応するプログラムを実装せよ． **

** 30 形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．ただし，各形態素は表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとするマッピング型に格納し，1文を形態素（マッピング型）のリストとして表現せよ．第4章の残りの問題では，ここで作ったプログラムを活用せよ **

mecabの出力フォーマットは

表層形\t品詞,品詞細分類1,品詞細分類2,品詞細分類3,活用型,活用形,原形,読み,発音

http://taku910.github.io/mecab/#usage-tools

従って、
- surface: 0
- base : 7
- pos (parts of speech, 品詞): 1
- pos1 : 2


```python
type([])
```

```python
run_decomposed = []
with codecs.open(fname_parsed, 'r' , 'utf-8') as f0:
    for x in f0:
        run_split = x.replace('\t',',').split(',')
        if len(run_split) > 7:
            run_map = {
                'surface':run_split[0],
                'base':run_split[7] ,
                'pos':run_split[1],
                'pos1':run_split[2]
            }
            run_decomposed.append(run_map)
        #print(x.replace('\t'))
```

#### 動詞の表層形をすべて抽出せよ．

```python
for x in run_decomposed:
    if x['pos'] == '動詞':
        print(x['surface'])
```

#### 動詞の原形をすべて抽出せよ．

```python
for x in run_decomposed:
    if x['pos'] == '動詞':
        print(x['base'])
```

#### サ変接続の名詞をすべて抽出せよ．

```python
for x in run_decomposed:
    if x['pos1'] == 'サ変接続':
        print(x['surfaceo
```

##### サ変接続の名詞とは
サ変接続というのは「〜する」の形が取れるものです。そのため品詞が名詞で、品詞細分類1がサ変接続のものを探せば良いようです。
https://qiita.com/segavvy/items/229d0a672a866b3e09a1

```python
for x in run_decomposed:
    if (x['pos1'] == 'サ変接続') and (x['pos'] != '名詞'):

        print('{0} : {1}'.format(x['pos'] , x['pos1']))
```

##### sandbox

#### 2つの名詞が「の」で連結されている名詞句を抽出せよ．

```python
for i in range(2, len(run_decomposed)):
    check0 = (run_decomposed[i-2]['pos'] == '名詞')
    check1 = run_decomposed[i-1]['surface'] == 'の'
    check2 = run_decomposed[i]['pos'] == '名詞'
    if check0 and check1 and check2:
        print('{0}{1}{2}'.format(run_decomposed[i-2]['surface'] , run_decomposed[i-1]['surface'] , run_decomposed[i]['surface']))
```

#### 名詞の連接（連続して出現する名詞）を最長一致で抽出せよ

```python
run_flags = ""
for i in range(len(run_decomposed)):
    if run_decomposed[i]['pos'] == '名詞':
        run_flags += "A"
    else:
        run_flags += "B"
```

正規表現に関して：

http://uxmilk.jp/41416

```python
import re
```

```python
run_pattern = re.compile('[^B](A{2,})[$B]')
```

```python
for x in run_pattern.finditer(run_flags):
    print("".join([run_decomposed[x]['surface'] for x in range(x.span(1)[0] , x.span(1)[1])]))
    print(x.group(1))
```

#### 文章中に出現する単語とその出現頻度を求め，出現頻度の高い順に並べよ．

```python
from collections import Counter
```

```python
word_counter = Counter()
```

```python
for x in run_decomposed:
    word_counter.update([x['surface']])
```

```python
word_counter.most_common()
```

#### 出現頻度が高い10語とその出現頻度をグラフ（例えば棒グラフなど）で表示せよ．

```python
import plotly.graph_objs as go
import plotly.offline as offline
offline.init_notebook_mode()
```

```python
commons = word_counter.most_common()
```

```python
commons[10
       ][0]
```

```python
common_words = [commons[i][0] for  i  in range(10)]
```

```python
common_word_counts = [commons[i][1] for  i  in range(10)]
```

```python
offline.iplot([go.Bar(x = common_words , y = common_word_counts )])
```

#### 単語の出現頻度順位を横軸，その出現頻度を縦軸として，両対数グラフをプロットせよ．
