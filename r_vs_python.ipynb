## データ選択

使える引数の形式	ix	iloc	loc

index	○	-	○

順序　(順序のリスト)	○	○	-

index, もしくは columns と同じ長さの bool のリスト	○	○	○

- locはインデックスを引数として取る（インデックスではなく「名前」と覚えるのがいいか・・）
- ilocは位置を引数として取る (iは位置とでも覚えるのがよいか・・）
 - ilocで最後
- 列を全部とる場合単に「:」でOK
- ilocは最後を切り取るときなどにiloc[-1]などとできる

```python
flights.iloc[-1]
```

locはインデックスを取るので、以下はエラーになる

```python
flights[flights.month == 2].loc[1,]
```

ilocは位置なので以下がエラーにならない

```python
flights[flights.month == 2].iloc[1,]
```

列に関してのみ、以下のようにして指定できる

```python
flights['year']
```

行を名前で指定する場合はlocを使う：

```python
flights.loc[1]
```

以下はエラーになる

```python
flights[1]
```
