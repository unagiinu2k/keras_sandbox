# データ選択

## ilocはintegerでアクセス。locは名前でアクセス

使える引数の形式	ix	iloc	loc

index	○	-	○

順序　(順序のリスト)	○	○	-

index, もしくは columns と同じ長さの bool のリスト	○	○	○

- locはインデックスを引数として取る（インデックスではなく「名前」と覚えるのがいいか・・）
- ilocは位置を引数として取る (iは位置あるいはintegerとでも覚えるのがよいか・・）
 - ilocで最後
- 列を全部とる場合単に「:」でOK
- ilocは最後を切り取るときなどにiloc[-1]などとできる

```python
flights.iloc[-1]
```

locはインデックスを取るので、以下はエラーになる

```python
flights[flights.month == 2].loc[1,]
```

ilocは位置なので以下がエラーにならない

```python
flights[flights.month == 2].iloc[1,]
```

## 列選択は[]

列に関してのみ、以下のようにして指定できる

```python
flights['year']
```

## 行選択はloc[]
行を名前で指定する場合はlocを使う：

```python
flights.loc[1]
```

以下はエラーになる

```python
flights[1]
```

locに列名だけ指定、というのエラーになる:

```python
flights.loc['year']
```

以下のようにすればよい

```python
flights.loc[:,'year']
```

## 列名と行名が被る場合の実験

- df[]は常に列選択 
- df.loc[],df.iloc[]に第一引数のみを指定した場合は常に行選択

ということでこれらを使っている限りでは列名と行名の衝突は問題にはならなそう

```python
tmp = flights.iloc[0:2,0:2].copy()
```

```python
tmp.columns = ['a' , 'b']
```

```python
tmp.index = ['a', 'b']
```

```python
tmp.loc['a']
```

```python
tmp['b']
```

```python
tmp
```

```python
tmp.loc['year']
```

## summarize
dplyrのsummarizeはaggで実現される

```python
flights.groupby('day').agg(dict(month = "mean"))
```

以下はシステマティックだと思うがワーニングがでる：

```python
flights.groupby('day').agg(dict(month = dict(mean  = "mean" , median = "median")))
```

名前の自由度は低いが以下のようにするのが正しいらしい：
https://stackoverflow.com/questions/44635626/pandas-aggregation-warning-futurewarning-using-a-dict-with-renaming-is-depreca

```python
flights.groupby('day').agg(dict(month = ["mean" , "median"]))
```

aggした後は普通のDataFrameになっている：

```python
type(flights.groupby('day').agg(dict(month = ["mean" , "median"])))
```

## rolling

```python
flights.groupby(['year' , 'day']).rolling(3).mean()
```

```python
flights.groupby(['day'])['month'].rolling(3).mean()
```

```python
roll_mean = flights.groupby(['day'])['month'].rolling(3).mean()
```

```python
type(roll_mean)
```

### pandas.core.series.Seriesを数値リストにするには

```python
roll_mean.values
```

### groupby, roll_mean, mutateに相当することをするには

```python
flights.assign(tmp = roll_mean.values)
```

```python
flights.assign(tmp = flights.groupby(['day'])['month'].rolling(3).mean().values)
```
